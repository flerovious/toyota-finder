### Data extraction with langchain

```python
from langchain_core.pydantic_v1 import BaseModel, Field
from typing import Optional

class Car(BaseModel):
    make: str
    model: str
    year: int
    description: str
    review: str

class CarScore(BaseModel):
    make: str
    model: str
    year: int
    risk_profile_score: Optional[float] = Field(None, description="Score for risk profile based on horsepower and speed preferences")
    social_profile_score: Optional[float] = Field(None, description="Score for social usage like family or friends transportation")
    capacity_score: Optional[float] = Field(None, description="Score for the capacity to carry people")
    curiosity_off_road_score: Optional[float] = Field(None, description="Score for off-road capability and curiosity to explore new places")
    moral_score: Optional[float] = Field(None, description="Score for preference towards electric vs gas vehicles")
    mpg_efficiency_score: Optional[float] = Field(None, description="Score for fuel efficiency")
    safety_score: Optional[float] = Field(None, description="Score for safety features")
    year_preference_score: Optional[float] = Field(None, description="Score based on how old the car is")
```

```python
from langchain_openai import ChatOpenAI

model = ChatOpenAI(model="gpt-4-turbo", temperature=0)

car_llm = model.with_structured_output(Car)
```

```python
car_llm.invoke(f"You are a professional web scraper. Extract relevant information from the page given the html. Here is the html: {review_section_html}")
```

```python
car_score_llm = model.with_structured_output(CarScore)
```

```python
car_score_llm.invoke(f"You are a car expert working on building a car recommendation engine. You are to give a score for various metrics for better recommendation. Here is the html of the page about the car: {review_section_html}")
```
